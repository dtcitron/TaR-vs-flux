```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

# Fast Travel, High Beta_2

```python
# Reading in some data:

data_path = "Data/SIR_Fast_high.csv"
fast_high = np.zeros((301,3))

with open(data_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
            fast_high[line_count] = row
            line_count += 1
    print(f'Processed {line_count} lines.')
    

fast_high = fast_high.transpose()
```

```python
plt.figure(figsize =(8,6))

plt.plot(fast_high[0], fast_high[1]/500, linewidth = 8, color = "b")
plt.plot(fast_high[0], fast_high[2]/500, linewidth = 8, color = "r")

plt.ylim(0,0.9)
# axes label size
plt.tick_params(axis = "both", which = "major", labelsize = 24)
plt.xlabel(r"Transmission Intensity $\beta_1/\gamma$", fontsize = 24)
plt.ylabel(r"$S^\infty_1/N_1$", fontsize = 24, rotation = 90)
plt.xticks([0,1,2,3], ["0","1","2","3"])
plt.yticks([0,.2,.4,.6,.8,1])#, ["0","1","2","3"])

# legend
plt.legend(["Simple Trip Model","Flux Model"], \
           loc = ("upper right"), fontsize = 24)

plt.tight_layout()

plt.savefig("Figures/SIR_Fast_high_patch1.pdf")
```

```python
data_path = "Data/SIR_Fast_high_time_series.csv"
fast_high_ts = np.zeros((241,5))

with open(data_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
            fast_high_ts[line_count] = row
            line_count += 1
    print(f'Processed {line_count} lines.')
    

fast_high_ts = fast_high_ts.transpose()
```

```python
plt.figure(figsize =(8,6))

plt.plot(fast_high_ts[0], fast_high_ts[1]/500, linewidth = 8, color = "b")
plt.plot(fast_high_ts[0], fast_high_ts[2]/500, linewidth = 8, color = "r")

# axes label size
plt.tick_params(axis = "both", which = "major", labelsize = 24)
plt.xlabel(r"time $t$", fontsize = 24)
plt.ylabel(r"$S_1(t)/N_1$", fontsize = 24, rotation = 90)
plt.xticks([0,10,20,30])
plt.yticks([0,.2,.4,.6,.8,1])#, ["0","1","2","3"])
plt.xlim(0,30)
plt.ylim(0,1.)

# legend
plt.legend(["Simple Trip Model","Flux Model"], \
           loc = ("upper right"), fontsize = 24)

plt.tight_layout()

plt.savefig("Figures/SIR_Fast_high_patch1_ts.pdf")
```
