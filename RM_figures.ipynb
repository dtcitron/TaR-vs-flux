```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

# Fast Travel, High PR_2

```python
# Reading in some data:

data_path = "Data/RM_Fast_high.csv"
fast_high = np.zeros((500,3))

with open(data_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
            fast_high[line_count] = row
            line_count += 1
    print(f'Processed {line_count} lines.')
    

fast_high = fast_high.transpose()
```

```python
plt.figure(figsize =(8,6))

plt.plot(fast_high[1], fast_high[0], linewidth = 8, color = "b")
plt.plot(fast_high[2], fast_high[0], linewidth = 8, color = "r")
#plt.plot(np.arange(-.1,.5,.01),np.array([0]*60), "k--", linewidth = 2)

plt.ylim(-.01,.5)
plt.xlim(-0.5,3)
# axes label size
plt.tick_params(axis = "both", which = "major", labelsize = 24)
plt.xlabel(r"$R_0$, Location 1", fontsize = 24)
plt.ylabel(r"Prevalence  $X_1/N_1$", fontsize = 24)
plt.yticks([0,.1,.2,.3,.4,.5])
plt.xticks([0,1,2,3])

plt.legend(["Simple Trip Model","Flux Model"], \
           loc = ("upper left"), fontsize = 24)

plt.tight_layout()

plt.savefig("Figures/RM_Fast_high_patch1.pdf")
```
